```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import ticker
import math
from typing import Optional, List
from sortedcontainers import SortedDict
from Plasma import Plasma_Evolver
from scipy import interpolate
import itertools
import pickle
```

```python
# Data for the first graph
x1 = np.linspace(0, 1, 100)
x2 = np.linspace(0, 1, 21)

passive = x2[::2]
active = x2[1::2]


# Create a figure and two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(10,5), dpi=150)

# Plot the first graph
ax2.scatter(passive, np.sin(2*np.pi*passive), c='b', s=20)
ax2.scatter(active, np.sin(2*np.pi*active), c='r', s=20)
ax2.plot(x1, np.sin(2*np.pi*x1), c='k', linewidth=0.5, linestyle='--')
ax2.set_xticks([])  # Remove x-axis ticks
ax2.set_yticks([])  # Remove y-axis ticks


# Plot the second graph
ax1.plot(x1, np.sin(2*np.pi*x1), c='k', linewidth=0.5, linestyle='--')
ax1.scatter(active, np.sin(2*np.pi*active), c='r', s=20)
ax1.scatter(passive, np.sin(2*np.pi*passive), c='k', marker ='|', s=15)
ax1.set_xticks([])  # Remove x-axis ticks
ax1.set_yticks([])  # Remove y-axis ticks


# Adjust the spacing between subplots
plt.subplots_adjust(hspace=0)

# Display the plot
plt.show()
```

```python
x = np.linspace(-1 ,1, 1000)

f = lambda x: -x ** 2 + 0.4 * x


fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10,4), dpi=150)

ax1.plot(x, f(x), c='k', linewidth=0.5, linestyle='--')
ax1.scatter([-1, 1], [f(-1), f(1)], c='b')
ax1.scatter(0.1, f(0.1), c='r')

d = 0.05

ax1.plot([-1 + d, 1 - d], [f(-1) + d*0.5, f(1) - d*0.5], linestyle='-', linewidth=0.5, c='k')

ax1.text(0, -1.1, r"$d_1$")


ax1.text(-0.95, -1.42, r"$p_1$")
ax2.text(-0.95, -1.42, r"$p_1$")

ax1.text(0.06,-0.06,r"$p_2$")
ax2.text(0.06,-0.06,r"$p_2$")

ax1.text(0.95,-0.7,r"$p_3$")
ax2.text(0.95,-0.7,r"$p_3$")

ax2.text(-0.48, -0.48, r"$p_4$")
ax2.text(0.52, -0.18, r"$p_5$")


ax2.plot(x, f(x), c='k', linewidth=0.5, linestyle='--')
ax2.scatter([-1, 0.1, 1], [f(-1), f(0.1), f(1)], c='b')
ax2.scatter([-0.45, 0.55], [f(-0.45), f(0.55)], c='r')


ax1.set_xticks([])  # Remove x-axis ticks
ax1.set_yticks([])  # Remove y-axis ticks

ax2.set_xticks([])  # Remove x-axis ticks
ax2.set_yticks([])  # Remove y-axis ticks

plt.subplots_adjust(wspace=0)

plt.show()
```
